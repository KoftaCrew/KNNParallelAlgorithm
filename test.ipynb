```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open("output.txt", "r")
cluster = -1
xs = []
ys = []
c = []
for line in f:
  line = str(line)
  if line.startswith("Cluster"):
    cluster += 1
  else:
    x, y = line.split(", ")
    x = x[1::]
    y = y[:-2:]
    xs.append(x)
    ys.append(y)
    c.append(cluster)
f.close()
```

```python
for x in c:
    print(x)
```

```python
figure = plt.figure()
ax = figure.add_axes([0, 0, 1, 1])
ax.set_yticks([0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
ax.set_xticks([0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
ax.scatter(xs, ys, c=c)
figure.show();
```
